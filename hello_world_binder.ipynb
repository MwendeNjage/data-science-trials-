```python
import pandas as pd 
import numpy as np
import matplotlib.pylab as plt
d=pd.read_csv(r'C:\Users\HP\OneDrive\Desktop\olympics medals 2021\Tokyo 2021 dataset v3.csv')
df=pd.DataFrame(d)
print(d)
```

```python
# Sort the DataFrame by 'Total' column in descending order
df_sorted = df.sort_values(by='Total', ascending=False)

# Select the top ten countries
top_ten = df_sorted.head(10)

# Set 'Team/NOC' as the index for better visualization
top_ten.set_index('Team/NOC', inplace=True)

# Select relevant columns for visualization
medals_df = top_ten[['Gold Medal', 'Silver Medal', 'Bronze Medal']]

# Plot the medal distribution for each country
ax = medals_df.plot(kind='bar', figsize=(12, 8), colormap='viridis')

# Add labels and title
plt.title('Medal Distribution by Country (Top Ten)')
plt.xlabel('Country')
plt.ylabel('Number of Medals')

# Add legend
plt.legend(title='Medal Type')

# Rotate x-axis labels for better readability
plt.xticks(rotation=45)

# Show the plot
plt.show()
```

```python
# Sort the DataFrame by 'Total' column in descending order
df_sorted = df.sort_values(by='Total', ascending=False)

# Select the top ten countries
bottom_ten = df_sorted.tail(10)

# Set 'Team/NOC' as the index for better visualization
bottom_ten.set_index('Team/NOC', inplace=True)

# Select relevant columns for visualization
medals_df = bottom_ten[['Gold Medal', 'Silver Medal', 'Bronze Medal']]

# Plot the medal distribution for each country
ax = medals_df.plot(kind='bar', figsize=(12, 8), colormap='viridis')

# Add labels and title
plt.title('Medal Distribution by Country (Bottom Ten)')
plt.xlabel('Country')
plt.ylabel('Number of Medals')

# Add legend
plt.legend(title='Medal Type')

# Rotate x-axis labels for better readability
plt.xticks(rotation=45)

# Show the plot
plt.show()
```

```python
#Plot histograms 
plt.figure(figsize=(10,6))
#Histogram for Gold Medals
plt.hist(df['Gold Medal'], bins=20, color='gold', alpha=0.7, label='Gold Medal ')
# Histogram for Silver Medals
plt.hist(df['Silver Medal'], bins=20, color='silver', alpha=0.7, label='Silver Medal')
# Histogram for Bronze Medals
plt.hist(df['Bronze Medal'], bins=20, color='#cd7f32', alpha=0.7, label='Bronze Medal')
#Add labels and titles
plt.title('Distribution of Medals')
plt.xlabel('Number of Medals')
plt.ylabel('Frequency')
#Add Legend
plt.legend()
#show the plot
plt.show()
```

```python
#plotting density plot
ax=df['Silver Medal'].plot(kind='kde', alpha=0.9, color='blue', label= 'Silver medal')
ax=df['Bronze Medal'].plot(kind='kde', color='green',label='Bronze Medal')
ax=df['Gold Medal'].plot(kind='kde',color='gold',label='Gold Medal')
plt.title('Distribution of Medals')
plt.xlabel('Number of Medals')
plt.ylabel('Frequency')
#Add Legend
plt.legend()
#show the plot
plt.show()
```

```python
# Set 'Team/NOC' as the index for better visualization
df.set_index('Team/NOC')

# Select relevant columns for visualization
medals_df = df[['Gold Medal', 'Silver Medal', 'Bronze Medal']]

# Plot density for each medal type
plt.figure(figsize=(10, 6))

for medal in medals_df.columns:
    ax = df[medal].plot(kind='kde', label=f'{medal} Density')

# Add labels and title
plt.title('Density Plot of Medals by Team/NOC')
plt.xlabel('Number of Medals')
plt.ylabel('Density')

# Add legend
plt.legend()

# Show the plot
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pylab as plt
d=pd.read_csv(r'C:\Users\HP\OneDrive\Desktop\olympics medals 2021\Tokyo 2021 dataset v3.csv')
df=pd.DataFrame(d)
print(d)

# Sort the DataFrame by 'Total' column in descending order
df_sorted = df.sort_values(by='Total', ascending=False)

# Select the top ten countries
top_ten = df_sorted.head(10)
sns.pairplot(top_ten,vars=['Total',  'Rank by Total', 'NOCCode','Continent'])
plt.show()
```

```python
# Select the specified columns
columns_to_plot = ["Rank", "Gold Medal", "Silver Medal", "Bronze Medal", "Total", "Rank by Total", "NOCCode", "Continent"]

# Create a pair plot
sns.pairplot(df[columns_to_plot], hue='Continent', diag_kind='kde')

# Add titles and adjust layout
plt.suptitle('Pair Plot of Rank, Gold Medal, Silver Medal, Bronze Medal, Total, Rank by Total, NOCCode, and Continent', y=1.02)
plt.tight_layout()

# Show the plot
plt.show()
```

```python

df_corr=df[['Rank', 'Gold Medal', 'Silver Medal', 'Bronze Medal', 'Total', 'Rank by Total', 'NOCCode','Continent']].dropna().corr()
print(df_corr)
sns.heatmap(df_corr, annot=True)
```
